```python
import ipywidgets as widgets
from IPython.display import display
import json

# Save settings to a JSON file
def save_settings(button):
    settings = {}
    
    for phase_index, phase_box in enumerate(phases_box.children):
        phase_settings = {}
        number_balls = phase_box.children[1].value
        phase_settings["number_balls"] = number_balls
        phase_settings["balls"] = []

        ball_settings_box = ball_settings_boxes[phase_index]
        for ball_box in ball_settings_box.children:
            ball_settings = {
                "radius": ball_box.children[1].value,
                "speed": ball_box.children[2].value,
                "points_per_reinforcement": ball_box.children[3].value,
                "change_to_clicks": ball_box.children[4].value,
                "change_to_delay": ball_box.children[5].value,
                "color": ball_box.children[6].value
            }
            phase_settings["balls"].append(ball_settings)
        
        settings[f"phase_{phase_index + 1}"] = phase_settings

    with open("settings.json", "w") as f:
        json.dump(settings, f)
    print("Settings saved to settings.json")

# Load settings from a JSON file
def load_settings(button):
    try:
        with open("settings.json", "r") as f:
            settings = json.load(f)
        
        # Set number of phases
        num_phases = len(settings)
        number_phases_widget.value = num_phases
        
        # Adjust the phases_box to match the loaded number of phases
        phases_box.children = []
        ball_settings_boxes.clear()  # Clear previous ball settings boxes

        # Set number of balls and their respective settings for each phase
        for phase_index, (phase_name, phase_data) in enumerate(settings.items()):
            num_balls = phase_data["number_balls"]

            # Create a new phase label and number of balls widget
            phase_label = widgets.Label(f"Phase {phase_index + 1}")
            number_balls_phase_widget = widgets.BoundedIntText(
                value=num_balls, min=1, max=7, description="Number of balls:", layout=widgets.Layout(width='160px')
            )
            
            # Add a box for the current phase's ball settings
            ball_settings_box = widgets.VBox([])

            # Update the ball settings based on loaded data
            for ball_index in range(num_balls):
                ball_data = phase_data["balls"][ball_index]

                # Dynamic widgets for ball options
                radii_widget = widgets.IntSlider(value=ball_data["radius"], min=10, max=120, step=10, description="Radius:", layout=widgets.Layout(width='250px'))
                initial_speed_widget = widgets.FloatSlider(value=ball_data["speed"], min=0.5, max=4.0, step=0.1, description="Speed:", layout=widgets.Layout(width='250px'))

                # Points per reinforcement
                points_per_reinforcement_widget = widgets.IntText(value=ball_data["points_per_reinforcement"], description="Points:", layout=widgets.Layout(width='140px'))

                # Change to clicks
                change_to_clicks_widget = widgets.IntText(value=ball_data["change_to_clicks"], description="Clicks:", layout=widgets.Layout(width='140px'))

                # Change to delay
                change_to_delay_widget = widgets.IntText(value=ball_data["change_to_delay"], description="Delay:", layout=widgets.Layout(width='140px'))

                # Ball color selection
                color_widget = widgets.ColorPicker(value=ball_data["color"], description="Color:", layout=widgets.Layout(width='250px'))

                # Arrange all widgets for the current ball in a horizontal row (HBox)
                ball_row = widgets.HBox([
                    widgets.Label(f"Ball {ball_index + 1}:", layout=widgets.Layout(width='70px', padding='0 5px 0 0')),  # Reduced padding
                    radii_widget,
                    initial_speed_widget,
                    points_per_reinforcement_widget,
                    change_to_clicks_widget,
                    change_to_delay_widget,
                    color_widget
                ])

                # Add the row for the current ball to the ball settings box
                ball_settings_box.children += (ball_row,)

            # Add the phase widgets to the phase container
            phases_box.children += (
                widgets.VBox([
                    phase_label, 
                    number_balls_phase_widget, 
                    ball_settings_box
                ]),
            )

            # Store the ball settings box for the current phase
            ball_settings_boxes.append(ball_settings_box)

            # Observe the number of balls for the current phase
            number_balls_phase_widget.observe(lambda change, phase=phase_index: update_balls_fields(phase, change['new']), names='value')
        
        print("Settings loaded from settings.json")
    
    except FileNotFoundError:
        print("No settings file found. Please save settings first.")
    except Exception as e:
        print(f"Error loading settings: {e}")

# Function to update dynamic fields based on number of phases and balls
def update_dynamic_fields(change):
    num_phases = number_phases_widget.value
    phases_box.children = []  # Clear previous phases
    ball_settings_boxes.clear()  # Clear previous ball settings boxes

    # Create dynamic widgets for each phase
    for phase in range(num_phases):
        phase_label = widgets.Label(f"Phase {phase + 1}")
        
        # Number of balls for this phase
        number_balls_phase_widget = widgets.BoundedIntText(
            value=3, min=1, max=7, description="Number of balls:", layout=widgets.Layout(width='160px')  # Wider input
        )
        
        # Function to update ball-specific fields when the number of balls changes
        def update_balls_fields(phase_index, num_balls):
            ball_settings_box = ball_settings_boxes[phase_index]
            ball_settings_box.children = []  # Clear previous ball settings

            # Ball-specific widgets for each ball in this phase
            for ball in range(num_balls):
                # Dynamic widgets for ball options
                radii_widget = widgets.IntSlider(value=60, min=10, max=120, step=10, description="Radius:", layout=widgets.Layout(width='250px'))  # Wider slider
                initial_speed_widget = widgets.FloatSlider(value=1.0, min=0.5, max=4.0, step=0.1, description="Speed:", layout=widgets.Layout(width='250px'))  # Wider slider

                # Points per reinforcement
                points_per_reinforcement_widget = widgets.IntText(value=1, description="Points:", layout=widgets.Layout(width='140px'))  # Wider input

                # Change to clicks
                change_to_clicks_widget = widgets.IntText(value=1, description="Clicks:", layout=widgets.Layout(width='140px'))  # Wider input

                # Change to delay
                change_to_delay_widget = widgets.IntText(value=1, description="Delay:", layout=widgets.Layout(width='140px'))  # Wider input

                # Ball color selection
                color_widget = widgets.ColorPicker(value="#ff0000", description="Color:", layout=widgets.Layout(width='250px'))  # Wider color picker

                # Arrange all widgets for the current ball in a horizontal row (HBox)
                ball_row = widgets.HBox([
                    widgets.Label(f"Ball {ball + 1}:", layout=widgets.Layout(width='70px', padding='0 5px 0 0')),  # Reduced padding
                    radii_widget,
                    initial_speed_widget,
                    points_per_reinforcement_widget,
                    change_to_clicks_widget,
                    change_to_delay_widget,
                    color_widget
                ])

                # Add the row for the current ball to the ball settings box
                ball_settings_box.children += (ball_row,)

        # Box to hold settings for each ball in the current phase
        ball_settings_boxes.append(widgets.VBox([]))

        # Observe the number of balls for the current phase
        number_balls_phase_widget.observe(lambda change, phase=phase: update_balls_fields(phase, change['new']), names='value')
        
        # Initial setup for the first ball configuration in the phase
        update_balls_fields(phase, number_balls_phase_widget.value)

        # Add phase widgets to the phase container
        phases_box.children += (
            widgets.VBox([
                phase_label, 
                number_balls_phase_widget, 
                ball_settings_boxes[phase]
            ]),
        )

# Widget to select the number of phases
number_phases_widget = widgets.BoundedIntText(
    value=1, min=1, max=5, description="Number of phases:", layout=widgets.Layout(width='160px')  # Wider input
)
number_phases_widget.observe(update_dynamic_fields, names='value')

# Initialize ball settings container for each phase
ball_settings_boxes = []

# Container for dynamically created phase inputs
phases_box = widgets.VBox([])

# Buttons for saving and loading settings
save_button = widgets.Button(description="Save Settings")
save_button.on_click(save_settings)

open_button = widgets.Button(description="Open Settings")
open_button.on_click(load_settings)

# Initial call to update all dynamic fields based on the default values
update_dynamic_fields(None)

# Display the widgets
display(widgets.VBox([number_phases_widget, save_button, open_button, phases_box]))
```
